```python
import requests
from bs4 import BeautifulSoup
import pandas as pd       # using pandas for data structure 
import nltk               # nlp library
import os                 # import the os module 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib as plt
import numpy as np
```

```python
def days_in_month(year,month):
    if month in [1,3,5,7,8,10,12]:
        return 31
    elif month == 2:
        if (year % 4) == 0:
            return 29
        else:
            return 28
    else:
        return 30
    
    
```

```python
for year in range(2009, 2024):
    for month in range(1,13):
        for day in range(1, days_in_month(year,month) + 1):
            
            strmonth = str(month)
            strday = str(day)
            
            if (month < 10):
                strmonth = "0" + strmonth
                
            if (day < 10):
                strday = "0" + strday
            
            
            url = f"https://www.wsj.com/news/archive/{year}/{strmonth}/{day}"
            
            session = requests.session()
            session.headers.update({'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/64.0.3282.167 Chrome/64.0.3282.167 Safari/537.36'})

            info = session.get(url)
            
            print(f"{year}-{strmonth}-{strday}")
            
            f = open(f"wsj_data/data{year}-{strmonth}-{strday}.html", "w")
            f.write(info.text)
            f.close()
```

```python
stop = stopwords.words('english')
stop = set(stop)        # change list to set significantly increases the processing speed

fn = "LoughranMcDonald_MasterDictionary_2018.csv" # input file                                                                                     fn = 'f:\factiva\PythonCodes\LoughranMcDonald_MasterDictionary_2018.csv' # input file
df_Master = pd.read_csv(fn)
df_copy = df_Master.copy()
master_list = list(df_Master['Word'])    
master_list = [itr for itr in master_list if not isinstance(itr,float) ]  # drop the floating numbers, leave the strings
master_list = [itr.lower() for itr in master_list] # make all lowercase
master_list = set(master_list)
```

```python
# Helper function for tokenize_df
def filter_func(word):
    if (word in master_list) and (word not in stop):
        return True # keep
    else:
        return False # filter out

def tokenize_headline(headline):
    headline = headline.lower()    # need to turn to lower cases 
    lst = headline.split() # split into list of words
    lst = list(filter(filter_func, lst)) # filter it using filter() method
    
        
    return lst
```

```python
year = 2022
strmonth = "04"
strday = "12"

def get_sentiments(year,strmonth,strday):
    f = open(f"wsj_data/data{year}-{strmonth}-{strday}.html", "r")
    html_doc = f.read()
    f.close()
    soup = BeautifulSoup(html_doc, 'html.parser')
    
    lst = soup.find_all("article")
    
    res = []
    
    for headline in lst:
        the_headline = headline.get_text()[:-13]
        res = res + (tokenize_headline(the_headline))
        
    return res
        
    
        
get_sentiments(year,strmonth,strday)
```

```python
# We get the list of positive/negative words from df_Master
pos_lst = list(df_Master.loc[df_Master['Positive']>0,'Word'])
pos_lst = [itr.lower() for itr in pos_lst]

neg_lst = list(df_Master.loc[df_Master['Negative']>0,'Word'])
neg_lst = [itr.lower() for itr in neg_lst]
```

```python
date = []
poswords = []
negwords = []
totwords = []

for year in range(1998, 2024):
    for month in range(1,13):
        strmonth = str(month)            
        if (month < 10):
            strmonth = "0" + strmonth
            
        pos = 0
        neg = 0
        tot = 0
        
        print(year, strmonth)
        
        for day in range(1, days_in_month(year,month) + 1):

            strday = str(day)
                
            if (day < 10):
                strday = "0" + strday
                
            
            keywords = get_sentiments(year,strmonth,strday)
            
            for word in keywords:
                if word in pos_lst:
                    pos += 1
                if word in neg_lst:
                    neg += 1
                    
                tot += 1
            
        date.append(str(year) + "-" + strmonth)
        poswords.append(pos)
        negwords.append(neg)
        totwords.append(tot)
        
df = pd.DataFrame(data = {
    "date" : date,
    "Positive" : poswords,
    "Negative" : negwords,
    "Total" : totwords
})

df
```
