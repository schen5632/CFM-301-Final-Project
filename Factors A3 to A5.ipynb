```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
master_df = pd.read_csv("factor1and2.csv")
master_df = master_df.drop(master_df.columns[0], axis = 1)
master_df["ticker"] = master_df["ticker"].str[2:-1]
master_df
```

# Revisions in Analyst Earnings Forecasts

```python
meanest = pd.read_csv("meanest.csv")
meanest["monthid"] = 12*(meanest["STATPERS"].str[0:4].astype(float) - 1980) + meanest["STATPERS"].str[5:7].astype(float)
meanest
```

```python
price = pd.read_csv("a3_price.csv")
price["monthid"] = 12*(price["STATPERS"].str[0:4].astype(float) - 1980) + price["STATPERS"].str[5:7].astype(float)
price
```

```python
merged_info = pd.merge(meanest, price[["oftic","TICKER", "monthid", "PRICE"]], on=["oftic", "TICKER", "monthid"], how="inner")
merged_info
```

```python
groups = merged_info.groupby(["oftic", "TICKER"])

res = []

for i, group in groups:
    tmp = (group["MEANEST"] - group["MEANEST"].shift(1))/group["PRICE"].shift(1)
    res = res + (list(tmp.rolling(6).sum()))
    
    
merged_info["Re"] = pd.Series(res)
merged_info
```

```python
tickers = []
monthids = []
Res = []

groups = merged_info.groupby(['oftic', "monthid"])

for (tic,moid), group in groups:
    tickers.append(tic)
    monthids.append(moid)
    Res.append(group["Re"].mean())
    
    
newdf = pd.DataFrame(data = {
    "ticker" : tickers,
    "monthid" : monthids,
    "Res" : Res,
})

master_df = pd.merge(master_df, newdf, on = ["ticker", "monthid"], how="left")
master_df
```

```python
master_df = master_df.sort_values(by = ['cusip', 'monthid']).reset_index(drop=True)
groups = master_df.groupby("cusip")

R6s = []
R11s = []

for i, group in groups:
    past6 = list(group["RET"].rolling(6).sum().shift(2))
    past11 = list(group["RET"].rolling(11).sum().shift(2))
    R6s = R6s + past6
    R11s = R11s + past11
    
master_df["R6"] = pd.Series(R6s)
master_df["R11"] = pd.Series(R11s)
master_df
```

```python
master_df.to_csv("factorsA1_to_A5.csv")
```
