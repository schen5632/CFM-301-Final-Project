# Standardized Unexpected Earnings (Sue)

Sue denotes Standardized Unexpected Earnings, and is calculated as the change
in split-adjusted quarterly earnings per share (Compustat quarterly item EPSPXQ divided by item
AJEXQ) from its value four quarters ago divided by the standard deviation of this change in
quarterly earnings over the prior eight quarters (six quarters minimum). At the beginning of each
month t, we split all stocks into deciles based on their most recent
past Sue. Starting from 1972, we use Sue computed with quarterly earnings from the most recent quarterly earnings announcement dates (Compustat quarterly item RDQ). For a firm to enter our portfolio formation, we require the
end of the fiscal quarter that corresponds to its most recent Sue to be within six months prior to
the portfolio formation. We do so to exclude stale information on earnings. To avoid potentially
erroneous records, we also require the earnings announcement date to be after the corresponding
fiscal quarter end. Monthly portfolio returns are calculated, separately, for the current month t
(Sue1), from month t to t +5 (Sue6), and from month t to t +11 (Sue12). Holding periods longer
than one month like in Sue6 mean that for a given decile in each month there exist six subdeciles,
each of which is initiated in a different month in the prior 6-month period. We average the subdecile
returns as the monthly return of the Sue6 decile.

```python
data = pd.read_csv("a1data.csv")
data
```

```python
groups = data.groupby("cusip")

sues = []

for i, group in groups:
    divisions = group["epspxq"]/group["ajexq"]
    stds = divisions.rolling(8).std()
    
    sues = sues + list(divisions.shift(4)/stds)
    
data["Sue"] = pd.Series(sues)
data.head(200)
```

At each month $t$, we split stocks into deciles based on Sue. The corresponding value of Sue will be the first value of rdq (Quarterly Earnings Announcement) that is prior to the date.

```python
data["monthid"] = 12*(data["rdq"].str[0:4].astype(float) - 1980) + data["rdq"].str[5:7].astype(float)
data

master_df = pd.merge(master_df, data[["cusip", "monthid", "Sue"]], on = ["cusip", "monthid"], how = "left")
master_df
```

```python
# Sue stays the same until the next announcement

groups = master_df.groupby("cusip")

res = []

for i, df in groups:
    
    ans = list(df["Sue"])
    
    prevans = np.nan
    result = []
    
    for j in range(0, len(ans)):
        if np.isnan(ans[j]):
            result.append(prevans)
        else:
            prevans = ans[j]
            result.append(prevans)

    res = res + result

    
master_df["Sue"] = pd.Series(res)
master_df
```

# Cumulative Abnormal Returns Around Earnings Announcement Dates

We calculate cumulative abnormal stock return (Abr) around the latest quarterly earnings
announcement date (Compustat quarterly item RDQ) as 

$$ Abr_i = \sum_{d = -2}^{+1} r_{id} - r_{md}$$

in which $r_{id}$ is stock $i$'s return on day $d$ (with the earnings announced at day 0), and $r_{md}$ is the value- weighted market index return. We cumulate returns until one (trading) day after the announcement date to account for the 1-day-delayed reaction to earnings news.

```python
# Read in datatypes
dtype_dict = {        
    'iid': str,       
    'cusip': str,
}

full_sp1500 = pd.read_csv("sp1500daily.csv", dtype=dtype_dict)
full_sp1500["Market Cap"] = full_sp1500["prccd"]*full_sp1500["cshoc"]
```

We calculate the weighted return for the market at each date.

```python
rm = []

dates = []
rms = []

groups = full_sp1500.groupby("datadate")

for i, group in groups:
    total = group["Market Cap"].sum()
    rets = (group["trfd"] - 1)*group["Market Cap"]/total
    mkt_ret = rets.sum()
    
    dates.append(i)
    rms.append(mkt_ret)
    

weighted_mkt_return = pd.DataFrame(data={
    "datadate" : dates,
    "weighted_mkt_ret" : rms
})

weighted_mkt_return
```

```python
# Get potential dates for abr

full_sp1500 = pd.merge(full_sp1500, weighted_mkt_return, on='datadate', how='left')
```

```python
groups = full_sp1500.groupby("cusip")

ans = []

for cusip,group in groups:
    #print(len(group))
    #print(group)
    res = (group["trfd"] - group["weighted_mkt_ret"]).rolling(4).sum().shift(-1)
    #print(len(res))
    ans = ans + list(res)
    
    
full_sp1500["abr"] = pd.Series(ans)
full_sp1500
```

```python
# For each announcement date, and for each stock, get the corresponding abr

res = []

groups = full_sp1500.groupby("cusip")


for i, row in data.iterrows():
    resp_df = groups.get_group(row["cusip"])

    abr = resp_df[resp_df["datadate"] == row["rdq"]]["abr"]
    
    if len(abr) == 0:
        res.append(np.nan)
    else:
        res.append(abr.iloc[0])
    
master_df["abr"] = pd.Series(res)

```
